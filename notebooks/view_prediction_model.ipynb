```python
from pylearn2.scripts.predict import get_prediction
from pylearn2.utils import serial
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from skimage.filter import threshold_otsu
import theano
%matplotlib inline
import os
```

```python
model = serial.load("../models/dae.pkl")
```

```python
data = np.load(os.path.join(os.getenv("DATA_PATH"), "ilc", "images.npy"))
data[data > 0] = 1
data = data.reshape( (data.shape[0], 18*18*30) )
data = data.astype(np.float32)

X = model.get_input_space().make_theano_batch()
h = model.encode(X)
X_hat = model.decode(h)

f = theano.function([X], X_hat)
data_hat = f(data)
```

```python

params = [p.get_value() for p in model.get_params()]
vb, hb, W, Wprime = params
```

```python

def get_pos(d):
    for z in xrange(30):
        for y in xrange(18):
            for x in xrange(18):
                if d[z, y, x] >= 0.5:
                    posx.append(x)
                    posy.append(y)
                    posz.append(z)
    return posx, posy, posz

fig = plt.figure()
n = 4
k = 1
for i in range(n):
    f_id = np.random.randint(0, data.shape[0] - 1)
    
    d = data[f_id].reshape( (30, 18, 18) )
    d_hat = data_hat[f_id].reshape( (30, 18, 18) )
    posx, posy, posz, cols = [], [], [], []
    ax = fig.add_subplot(n, 2, k, projection='3d', alpha=1.)
    posx, posy, posz = get_pos(d)
    ax.scatter(posx, posz, posy, c='red', marker='o', s=10)
    ax.set_xlabel("x")
    ax.set_ylabel("z")
    ax.set_zlabel("y")
    
    ax = fig.add_subplot(n, 2, k + 1, projection='3d', alpha=1.)
    posx, posy, posz = get_pos(d_hat)
    ax.scatter(posx, posz, posy, c='red', marker='o', s=10)
    ax.set_xlabel("x")
    ax.set_ylabel("z")
    ax.set_zlabel("y")
    k += 2

plt.show()
```
