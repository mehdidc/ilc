```python
from pylearn2.scripts.predict import get_prediction
from pylearn2.utils import serial
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from skimage.filter import threshold_otsu
%matplotlib inline
```

```python
model = serial.load("../models/predict.pkl")
```

```python
data = np.load("../data/images_predict_angle.npy.npz")
x = data['arr_0']
y = data['arr_1']
x = 1.*(x>0).astype(np.float32)
x = x.reshape( (x.shape[0], np.prod(x.shape[1:]) ) )
```

```python
y_hat = get_prediction(model, x)
```

```python
W, b, _, _, _ = model.get_params()
W = W.get_value()
b = b.get_value()
activations = np.maximum(np.dot(x, W) + b, 0)
```

```python
# Visualization by mean of TOP activations for each hidden unit

fig = plt.figure()
n_s=2
n=n_s**2
top = 50
for i in range(n):
    f_id = np.random.randint(0, activations.shape[1] - 1)
    
    list_of_activations = activations[:, i].tolist()
    indexes = range(0, activations.shape[1])
    indexes = sorted(indexes, key=lambda i:list_of_activations[i], reverse=True)[0:top]
    centroid = x[indexes].mean(axis=0).reshape( (30, 18, 18) )
    posx, posy, posz, cols = [], [], [], []
    for z_ in xrange(30):
        for y_ in xrange(18):
            for x_ in xrange(18):
                if centroid[z_, y_, x_] >= 1./top:  
                    posx.append(x_)
                    posy.append(y_)
                    posz.append(z_)
                    cols.append(centroid[z_, y_, x_])
    ax = fig.add_subplot(n_s, n_s, i + 1, projection='3d', alpha=1.)
    ax.scatter(posx, posz, posy, c=cols, cmap='autumn_r', marker='o', s=50)
    ax.set_xlabel("x")
    ax.set_ylabel("z")
    ax.set_zlabel("y")

plt.show()
```
